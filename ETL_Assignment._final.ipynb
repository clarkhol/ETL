```python
# import dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Store the Wikipedia URL
us_wiki_url = "https://en.wikipedia.org/wiki/List_of_states_and_territories_of_the_United_States_by_population#cite_note-5"

# Use Pandas to scrape Wikipedia page for tables
us_wiki_tables = pd.read_html(us_wiki_url)

us_pop_df = us_wiki_tables[0]
us_pop_df.head()

```

```python
us_pop_df.columns
```

```python
#us_pop_df.iloc[:,:-1]
```

```python
pop_us = us_pop_df.iloc[:,[1,2,-3,-2]]
pop_us.columns = pop_us.columns.droplevel(level=1)
pop_us.drop(['Rank'], axis=1, inplace=True)
pop_us.columns=[ "State", "Population - 2019", "Population - 2010"]
df2 = pop_us
df2.head()
```

```python
# Store the Wikipedia URL
can_wiki_url = "https://en.wikipedia.org/wiki/Population_of_Canada_by_province_and_territory"

# Use Pandas to scrape Wikipedia page for tables
can_wiki_tables = pd.read_html(can_wiki_url)

# first table
can_pop_df = can_wiki_tables[0]
can_pop_df.head()
```

```python
can_pop_df.columns
```

```python
pop_can = can_pop_df.iloc[:-1,[1,2,-3]]
pop_can.columns = pop_can.columns.droplevel(level=1)
pop_can.columns=[ "Provinces", "Population - 2016", "Population - 2020"]
df4 = pop_can
df4.head()
```

```python
#Total cnergy consumed by state 
us_file_path3 = "Resources/Total_energy_consumed.csv"

# create the data frame 
us_total_energy_df = pd.read_csv(us_file_path3)
us_total_energy_df.columns=[ "State", "Total Energy Consumed - BTU"]
df1 = us_total_energy_df
df1.head()
```
