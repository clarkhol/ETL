```python
# import dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Store the Wikipedia URL
us_wiki_url = "https://en.wikipedia.org/wiki/List_of_states_and_territories_of_the_United_States_by_population#cite_note-5"

# Use Pandas to scrape Wikipedia page for tables
us_wiki_tables = pd.read_html(us_wiki_url)

us_pop_df = us_wiki_tables[0]
us_pop_df.head()
```

```python
# Drop multindex 
us_pop_df.columns = us_pop_df.columns.droplevel(level=1)
us_pop_df.head()
```

```python
# Drop unwanted column names
us_pop_df = us_pop_df.drop(columns = ['Rank', 'Rank','Change, 2010–2019', 'Change, 2010–2019','Total U.S. House of Representatives Seats',
       'Estimated population per electoral vote, 2019[note 1]','Census population per House seat', 'Census population per House seat', 'Percent of the total U.S. population, 2019[note 2]'])
us_pop_df
```

```python
# Rename Columns in US Population - first census population is July 2019, second is april 1st, 2010
us_pop_df = us_pop_df.rename(columns = {"Census population":"Population 2019"})
us_pop_df
```

```python
# Drop the last population 2019 column, it was actually population 2010. 
us_pop_df = us_pop_df.iloc[:,0:-1]
us_pop_df.head()
```

```python
# Store the Wikipedia URL
can_wiki_url = "https://en.wikipedia.org/wiki/Population_of_Canada_by_province_and_territory"

# Use Pandas to scrape Wikipedia page for tables
can_wiki_tables = pd.read_html(can_wiki_url)

# first table
can_pop_df = can_wiki_tables[0]
can_pop_df.head()
```

```python
# Canadian Population drop level 
can_pop_df.columns = can_pop_df.columns.droplevel(level=1)
can_pop_df.head()
```

```python
# Canada population drop unwanted columns
can_pop_df = can_pop_df.drop(columns=['2016 PopulationRank','Growth,2011–16', 'Land area(km2)', 'Populationdensity(per km2)','House ofCommons seats','House ofCommons seats','Growth, 2016–20'])
can_pop_df
```

```python
# Rename Columns for Canadian Population
canadian_pop = can_pop_df.rename(columns = {"Name[1]": "Province", "Population,2016 Census":"Population 2016","2020 population(Q1 est.)[2]": "Population 2020"})
canadian_pop
```

```python
# Rename 2020 population (Q1 est.)[2] column name 
pop_2020_change = canadian_pop.rename(columns = {"2020 population(Q1 est.)[2]": "Population 2020"})
pop_2020_change
```

```python
# Use iloc function to view the table we'd like: Province,Population 2016 Number,Population 2020 Number
#can_pop_df.iloc[:,:-1]
canadian_pop.iloc[:,[1,2,-2]]
canadian_pop
```

```python
us_file_path2 = "Resources/US_energies.csv"

# create the data frame 
us_energy2_df = pd.read_csv(us_file_path2)
us_energy2_df.head()
```

```python
# Drop unwanted columns 
us_energy2_df = us_energy2_df.drop(columns = ['urbanUtilityScalePV_GWh', 'urbanUtilityScalePV_GW',
       'urbanUtilityScalePV_km2', 'ruralUtilityScalePV_GWh',
       'ruralUtilityScalePV_GW', 'ruralUtilityScalePV_km2', 'rooftopPV_GWh',
       'rooftopPV_GW', 'CSP_GWh', 'CSP_GW', 'CSP_km2','onshoreWind_GW', 'onshoreWind_km2','offshoreWind_GW', 'offshoreWind_km2', 'biopowerSolid_GW', 'biopowerSolid_BDT','biopowerGaseous_GW', 'biopowerGaseous_Tonnes-CH4','geothermalHydrothermal_GW','EGSGeothermal_GWh', 'EGSGeothermal_GW','hydropower_GW', 'hydropower_countOfSites'])
us_energy2_df.head()
```

```python
#Rename columns
us_energy2_df = us_energy2_df.rename(columns = {"onshoreWind_GWh":"Onshore Wind GWH", "offshoreWind_GWh": "Offshore Wind GWH", "biopowerSolid_GWh":"Biopower Solid GWH",
       "biopowerGaseous_GWh":"Biopower Gases GWH", "geothermalHydrothermal_GWh": "Geothermal GWH", "hydropower_GWh":"Hydropower GWH"})

us_energy2_df.head()
```

```python
# Drop all NaN values 
us_energy2_df['Onshore Wind GWH'].fillna("0",inplace = True)
us_energy2_df['Offshore Wind GWH'].fillna("0",inplace = True)
us_energy2_df['Biopower Solid GWH'].fillna("0",inplace = True)
us_energy2_df['Biopower Gases GWH'].fillna("0",inplace = True)
us_energy2_df['Geothermal GWH'].fillna("0",inplace = True)
us_energy2_df['Hydropower GWH'].fillna("0",inplace = True)
us_energy2_df.head()
```
